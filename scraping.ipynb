### Langkah-langkah

- Start project : ketikan "scrapy startproject <project_name>" pada terminal, saya memberi nama game_catalog
- kemudian akan terbentuk folder baru bernama "game_catalog" beserta isinya (sudah dari scrapy)
- kemudian di dalam folder spider, kita membuat folder dengan format .py untuk melakukan crawling dengan scrapy
- untuk menjalankan kodenya bisa mengetikkan "Scrapy crawl <spider_name> -o <file_name>.csv"
- hasil akan otomatis beradi di dalam folder project

```python
# kode yang saya gunakan

import scrapy
import pandas as pd

class najlaSpider(scrapy.Spider):
    name = "najla"

    start_urls = ["https://store.playstation.com/en-id/category/05a2d027-cedc-4ac0-abeb-8fc26fec7180/"]

    def parse(self, response):
        titles = response.css('span.psw-t-body.psw-c-t-1.psw-t-truncate-2.psw-m-b-2::text').getall()
        prices = response.css('span.psw-m-r-3::text').getall()

        for title, price in zip(titles, prices):
            yield {
                "Nama Game": title.strip(),
                "Harga Game": price.strip()
            }
        
        #untuk melanjutkan ke halaman selanjutnya
        for i in range (1,17):
            next_url = f"https://store.playstation.com/en-id/category/05a2d027-cedc-4ac0-abeb-8fc26fec7180/{i}"
            yield scrapy.Request(next_url, callback=self.parse)
```
